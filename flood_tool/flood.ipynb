```python
import os

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import SGDClassifier
from sklearn import svm

from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
```

# Data Preprocessing

```python
data = pd.read_csv("resources/postcodes_sampled.csv")
```

```python
X = data[["easting","northing"]]
y = data['riskLabel']
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42) # Holdout
```

```python
features = ["easting","northing","riskLabel"]
data_train, data_test = train_test_split(data, test_size=.3)

data_train = pd.DataFrame(data_train, columns=features)
data_test = pd.DataFrame(data_test, columns=features)
```

```python
data_cv = data_train.sample(frac=.8)
```

```python
X_train = data_train.drop('riskLabel', axis=1)
X_test = data_test.drop('riskLabel', axis=1)
X_cv = data_cv.drop('riskLabel', axis=1)
y_train = data_train['riskLabel'].copy()
y_test = data_test['riskLabel'].copy()
y_cv = data_cv['riskLabel'].copy()
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)


scoring = ['precision_macro', 'recall']
cv_results = cross_validate(clf, X_train, y_train, scoring=scoring)
pd.DataFrame(cv_results).mean()
```

```python
y_pred = clf.predict(X_test)
accuracy_svc = clf.score(X_test, y_test)
# accuracy2 = accuracy_score(y_test, y_pred)
print(accuracy_svc)
```

# Random Forest

```python
forest_classifier = RandomForestClassifier()
params = {'criterion': ['gini', 'entropy'],
         'max_features': ['auto', 'sqrt', 'log2']
         }

rf_gs = GridSearchCV(forest_classifier, params, scoring="accuracy", cv=5, n_jobs=-1)
rf_gs.fit(X_cv, y_cv)
```

```python
print(rf_gs.best_params_)
print(rf_gs.best_estimator_)
sgd_best = rf_gs.best_estimator_
print(sgd_best.score(X_test, y_test))
```

```python
forest_classifier = RandomForestClassifier(criterion = 'gini')
forest_classifier.fit(X_train, y_train)
cv_results = cross_validate(forest_classifier, X_train, y_train, scoring=['accuracy'])
```

```python
y_pred = forest_classifier.predict(X_test)
accuracy_rf = forest_classifier.score(X_test, y_test)
# accuracy2 = accuracy_score(y_test, y_pred)
print(accuracy_rf)
```

# KNN classifier

```python
neigh = KNeighborsClassifier()
params = {'n_neighbors':[2,3,4,5,10,20,30,50],
         'weights': ['uniform','distance'],
         'algorithm': ['auto', 'ball_tree','kd_tree', 'brute']
         }


knn_gs = GridSearchCV(neigh, params, scoring="accuracy", cv=5, n_jobs=-1)
knn_gs.fit(X_cv, y_cv)
```

```python
print(knn_gs.best_params_)
print(knn_gs.best_estimator_)
sgd_best = knn_gs.best_estimator_
print(sgd_best.score(X_test, y_test))
```

```python
neigh = KNeighborsClassifier(n_neighbors=20)
neigh.fit(X_train, y_train)
cv_results = cross_validate(neigh, X_train, y_train, scoring=['accuracy'])
pd.DataFrame(cv_results).mean()
```

```python
y_pred = neigh.predict(X_test)
accuracy_neigh = neigh.score(X_test, y_test)
# accuracy2 = accuracy_score(y_test, y_pred)
print(accuracy_neigh)
```

# SGD

```python
svc_bis = SGDClassifier()
params = {'penalty':['l2', 'l1', 'elasticnet'],
         'alpha': [1/20,1/10],
         'fit_intercept': [True, False],
          'loss': ['hinge', 'log', 'modified_huber', 'squared_hinge', 'perceptron', 'squared_error', 'huber', 'epsilon_insensitive','squared_epsilon_insensitive']
         }

# , 'poly', 'rbf', 'sigmoid', 'precomputed'
scoring = {"AUC": "roc_auc", "Accuracy": 'accuracy', 'f1': 'f1', 'precision': 'precision'}
sgd_gs = GridSearchCV(svc_bis, params, scoring=scoring, cv=5, refit='Accuracy', n_jobs=-1)
sgd_gs.fit(X_cv, y_cv)
```

```python
print(sgd_gs.best_params_)
print(sgd_gs.best_estimator_)
sgd_best = sgd_gs.best_estimator_
print(sgd_best.score(X_test, y_test))
```

```python
svc_bis = SGDClassifier(loss='hinge', penalty='l1', alpha=1/20)
svc_bis.fit(X_train, y_train)
cv_results = cross_validate(svc_bis, X_train, y_train, scoring=['accuracy'])
pd.DataFrame(cv_results).mean()
```

```python
y_pred = svc_bis.predict(X_test)
accuracy_svc_bis = svc_bis.score(X_test, y_test)
accuracy2 = accuracy_score(y_test, y_pred)
print(accuracy_svc_bis)
```
